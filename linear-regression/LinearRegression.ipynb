```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import math
import time
```

```python
dat = pd.read_csv("wine.csv", sep=";")
```

```python
plt.scatter(dat.loc[:,'pH'], dat.loc[:, 'density'], s=4)

x_var = np.array(dat.loc[:,'pH'])
y_var = np.array(dat.loc[:, 'density'])
n = len(x_var)

params = np.polyfit(x=x_var, y=y_var, deg=1)

padding = (max(x_var) - min(x_var))/20
lin_space = np.linspace(min(x_var)-padding, max(x_var)+padding)

print (params)

plt.plot(lin_space, params[1] + lin_space * params[0])
```

```python
fig, axs = plt.subplots(5, figsize=(5, 10))
plt.tight_layout()

axs[0].scatter(dat.loc[:,'pH'], dat.loc[:, 'density'], s=4)
axs[0].plot(lin_space, params[1] + lin_space * params[0])

m = 0
b = 1
m_list = []
b_list = []
dm_list = []
db_list = []

i = 0
change = 1
step = 0.1
while change > 0.000001:
    i += 1
    dm = np.dot(m * x_var + b - y_var, x_var)/n
    db = np.sum(m * x_var + b - y_var)/n
    m -= step * dm
    b -= step * db
    change = abs(dm) + abs(db)
    
    m_list.append(m)
    b_list.append(b)
    dm_list.append(dm)
    db_list.append(db)

axs[0].plot(lin_space, b + lin_space * m)

axs[1].plot(m_list)
axs[1].hlines(y=params[0], xmin=0, xmax=i, color='r')
axs[2].plot(b_list)
axs[2].hlines(y=params[1], xmin=0, xmax=i, color='r')

axs[3].plot(dm_list[1:20])
axs[4].plot(db_list[1:20])

print (str(i) + " Iterations")
print (str(n) + " Data Points")
print (m, b)
```
